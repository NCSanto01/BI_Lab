```python
# Data management
# ------------------------------------------------------------------------------
import numpy as np
import pandas as pd

# Graphics
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt

# Preprocessing and modelling
# ------------------------------------------------------------------------------
from simpful import *


# Configuraci√≥n warnings
# ------------------------------------------------------------------------------
import warnings
#warnings.filterwarnings('once')
warnings.filterwarnings('ignore')
```

```python
# A simple fuzzy inference system for the tipping problem
# Create a fuzzy system object
FS = FuzzySystem()
```

```python
# Define fuzzy sets and linguistic variables
baja = FuzzySet(points=[[0.0, 1.0],  [25.0, 1.0], [35., 0.]], term="baja")
media = FuzzySet(points=[[15., 0.], [35., 1.], [55., 0.]], term="media")
alta = FuzzySet(points=[[35., 0.],  [45., 1.], [70., 1.]], term="alta")
FS.add_linguistic_variable("Temperatura", LinguisticVariable([baja, media, alta], concept="Temperatura"))


```

```python
FS.plot_variable('Temperatura')
```

```python
FS.produce_figure('Temperatura')
```

```python
# Define output crisp values
FS.set_crisp_output_value("R1", 700.+500*40)
FS.set_crisp_output_value("R2", 100.+200*40)
FS.set_crisp_output_value("R3", 100.+50*40)
```

```python
# Define fuzzy rules
R1 = "IF (Temperatura IS alta)  THEN (Velocidad IS R1)"
R2 = "IF (Temperatura IS media) THEN (Velocidad IS R2)"
R3 = "IF (Temperatura IS baja)  THEN (Velocidad IS R3)"
FS.add_rules([R1, R2, R3])

```

```python
# Set antecedents values
FS.set_variable("Temperatura", 40.0)
#FS.set_variable("Food", 8)
```

```python
# Perform Sugeno inference and print output
print(FS.Sugeno_inference(["Velocidad"]))
```
