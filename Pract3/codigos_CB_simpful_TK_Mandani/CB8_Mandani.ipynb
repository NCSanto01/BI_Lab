```python
# Data management
# ------------------------------------------------------------------------------
import numpy as np
import pandas as pd

# Graphics
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt

# Preprocessing and modelling
# ------------------------------------------------------------------------------
from simpful import *


# Configuraci√≥n warnings
# ------------------------------------------------------------------------------
import warnings
#warnings.filterwarnings('once')
warnings.filterwarnings('ignore')
```

```python
# Create a fuzzy system object
FS = FuzzySystem()
```

```python
# Define input fuzzy sets and linguistic variables
x1 = FuzzySet(points=[[0.0, 1.0],  [15.0, 1.0], [30., 0.]], term="x1")
x2 = FuzzySet(points=[[10., 0.], [30.0, 1.0], [50., 0.]], term="x2")
x3 = FuzzySet(points=[[30., 0.],  [40., 1.], [60., 1.]], term="x3")
FS.add_linguistic_variable("X", LinguisticVariable([x1, x2, x3], concept="X"))

y1 = FuzzySet(points=[[0.0, 1.0],  [4., 0.]], term="y1")
y2 = FuzzySet(points=[[2., 0.], [6.0, 1.0], [12., 0.]], term="y2")
y3 = FuzzySet(points=[[8., 0.], [12., 1.]], term="y3")
FS.add_linguistic_variable("Y", LinguisticVariable([y1, y2, y3], concept="Y"))

```

```python
FS.plot_variable('X')
```

```python
FS.plot_variable('Y')
```

```python
# Define output fuzzy sets and linguistic variable
z1 = FuzzySet(points=[[0.0, 1.0],  [2., 0.]], term="z1")
z2 = FuzzySet(points=[[1., 0.], [3.0, 1.0], [5., 0.]], term="z2")
z3 = FuzzySet(points=[[2., 0.], [5., 1.]], term="z3")
FS.add_linguistic_variable("Z", LinguisticVariable([z1, z2, z3], concept="Z"))

```

```python
FS.produce_figure()
```

```python
# Define fuzzy rules
R1 = "IF (X IS x1)  AND (Y IS y1) THEN (Z IS z1)"
R2 = "IF (X IS x1)  AND (Y IS y3) THEN (Z IS z2)"
R3 = "IF (X IS x2)  AND (Y IS y2) THEN (Z IS z2)"
R4 = "IF (X IS x2)  AND (Y IS y3) THEN (Z IS z3)"
R5 = "IF (X IS x3)  AND (Y IS y1) THEN (Z IS z1)"


FS.add_rules([R1, R2, R3, R4, R5])

```

```python

# Define facts observed
input_x=45.0
input_y=10.0


# Set antecedents values
FS.set_variable("X", input_x)
FS.set_variable("Y", input_y)
```

```python
# GRado de cumplimiento de reglas
FS.get_firing_strengths()
```

```python
# Perform Sugeno inference and print output
print(FS.Mamdani_inference(["Z"]))
```
