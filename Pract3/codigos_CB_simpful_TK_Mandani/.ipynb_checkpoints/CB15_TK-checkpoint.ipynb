```python
# Data management
# ------------------------------------------------------------------------------
import numpy as np
import pandas as pd

# Graphics
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt

# Preprocessing and modelling
# ------------------------------------------------------------------------------
from simpful import *


# Configuraci√≥n warnings
# ------------------------------------------------------------------------------
import warnings
#warnings.filterwarnings('once')
warnings.filterwarnings('ignore')
```

```python
# Create a fuzzy system object
FS = FuzzySystem()
```

```python
# Define fuzzy sets and linguistic variables
bajo = FuzzySet(points=[[0.0, 1.0],  [10.0, 1.0], [30., 0.]], term="bajo")
normal = FuzzySet(points=[[5., 0.], [40.0, 1.0], [50., 1.], [80., 0.]], term="normal")
alto = FuzzySet(points=[[60., 0.],  [90., 1.], [100., 1.]], term="alto")
FS.add_linguistic_variable("T1", LinguisticVariable([bajo, normal, alto], concept="T1"))
FS.add_linguistic_variable("T2", LinguisticVariable([bajo, normal, alto], concept="T2"))
FS.add_linguistic_variable("T3", LinguisticVariable([bajo, normal, alto], concept="T3"))
```

```python
FS.plot_variable('T1')
```

```python
FS.produce_figure()
```

```python
# Define facts observed
t1=15.0
t2=33.0
t3=75.0
```

```python
# Define output crisp values
FS.set_crisp_output_value("R1", 0.4*t1+0.5*t2+0.7*t3)
FS.set_crisp_output_value("R2", 0.9*t1+0.8*t2+0.3*t3)
FS.set_crisp_output_value("R3", 0.1*t1+0.2*t2+0.3*t3)
FS.set_crisp_output_value("R4", 0.2*t1+0.1*t2+0.3*t3)
FS.set_crisp_output_value("R5", 0.7*t1+0.5*t2)
FS.set_crisp_output_value("R6", 0.1*t1+0.1*t2+0.2*t3)
FS.set_crisp_output_value("R7", 0.8*t1+0.7*t2+0.9*t3)
FS.set_crisp_output_value("R8", 0.8*t1+0.2*t2+0.1*t3)
FS.set_crisp_output_value("R9", 0.5*t1+0.5*t2+0.5*t3)
```

```python
# Define fuzzy rules
R1 = "IF (T1 IS bajo)  AND (T2 IS normal) AND (T3 IS normal) THEN (Potencia IS R1)"
R2 = "IF (T1 IS bajo)  AND (T2 IS normal) AND (T3 IS bajo) THEN (Potencia IS R2)"
R3 = "IF (T1 IS normal)  AND (T2 IS bajo) AND (T3 IS alto) THEN (Potencia IS R3)"
R4 = "IF (T1 IS bajo)  AND (T2 IS normal) AND (T3 IS alto) THEN (Potencia IS R4)" 
R5 = "IF (T1 IS normal)  AND (T2 IS alto) AND (T3 IS bajo) THEN (Potencia IS R5)"
R6 = "IF (T1 IS normal)  AND (T3 IS alto)  THEN (Potencia IS R6)"
R7 = "IF (T1 IS alto)  AND (T2 IS normal) AND (T3 IS alto) THEN (Potencia IS R7)" 
R8 = "IF (T1 IS bajo)  AND (T2 IS normal) THEN (Potencia IS R8)"
R9 = "IF (T1 IS alto)  AND (T3 IS bajo) THEN (Potencia IS R9)"

FS.add_rules([R1, R2, R3, R4, R5, R6, R7, R8, R9])

```

```python
# Set antecedents values
FS.set_variable("T1", t1)
FS.set_variable("T2", t2)
FS.set_variable("T3", t3)
```

```python
# GRado de cumplimiento de reglas
FS.get_firing_strengths()
```

```python
# Perform Sugeno inference and print output
print(FS.Sugeno_inference(["Potencia"]))
```
